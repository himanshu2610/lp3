```python
df.drop(columns = 'Email No.',inplace = True)
df
```

```python
df.isnull().any()
```

```python
df.Prediction.unique()
```

```python
df.Prediction = df.Prediction.replace({0 : 'Not Spam',1 : 'Spam'})
df
```

```python
X = df.drop(columns = 'Prediction',axis =1 )
Y = df.Prediction
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state = 1)
```

```python
KN = KNeighborsClassifier
knn = KN(n_neighbors = 7)
knn.fit(x_train,y_train)
y_pred_knn = knn.predict(x_test)
```

```python
print("Prediction: \n")
print(y_pred_knn)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score

cm_knn = confusion_matrix(y_test,y_pred_knn)
cm_knn
```

```python
print("Accuracy of the matrix is :",accuracy_score(y_test,y_pred_knn))
```

```python
from sklearn.svm import SVC
model = SVC(C = 1)
model.fit(x_train,y_train)
y_pred_svm = model.predict(x_test)
```

```python
cm_svm = confusion_matrix(y_test,y_pred_svm)
cm_svm
```

```python
print("Accuracy of the matrix is :",accuracy_score(y_test,y_pred_svm))
```
