```python
from math import *

def cal_distance(longitude1,latitude1,longitude2,latitude2):
  travel_dist = []
  for pos in range (len(longitude1)):
    lon1, lan1, lon2, lan2 = map(radians, [longitude1[pos], latitude1[pos], longitude2[pos], latitude2[pos]])
    dist_lon = lon2 - lon1
    dist_lan = lan2 - lan1

    a = sin(dist_lan/2)**2 + cos(lan1) * cos(lan2) * sin(dist_lon/2)**2

    #radius of earth = 6371
    c = 2 * asin(sqrt(a)) * 6371
    travel_dist.append(c)

  return travel_dist
```

```python
df["dist_travel_km"] = cal_distance(df["pickup_longitude"].to_numpy(),df["pickup_latitude"].to_numpy(),df["dropoff_longitude"].to_numpy(),df["dropoff_latitude"].to_numpy())
```

```python
def remove_outlier(df1,col):
  Q1 = df1[col].quantile(0.25)
  Q3 = df1[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_Whisker = Q1 - 1.5*IQR
  upper_Whisker = Q3 + 1.5*IQR
  df[col] = np.clip(df[col],lower_Whisker,upper_Whisker)
  return df1

def treat_outliers_all(df1,col_list):
  for c in col_list:
    df1 = remove_outlier(df,c)
  return df1
```

```python
df = treat_outliers_all(df,df.iloc[:,0::])
```

```python
df_x = df[['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count','hour','month','day','year','dayofweek','dist_travel_km']]
df_y = df['fare_amount']
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred_lin = reg.predict(x_test)
y_pred_lin
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 100)
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
y_pred_rf
```

```python
cols = ['Model','RMSE','R2_Score']

result = pd.DataFrame(columns = cols)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

reg_rmse = np.sqrt(mean_squared_error(y_test,y_pred_lin))
reg_squared = r2_score(y_test,y_pred_lin)

full_metrics = pd.Series({'Model': "Linear Regression", 'RMSE' : reg_rmse, 'R-Squared' : reg_squared})
full_metrics
```

```python
rf_RMSE = np.sqrt(mean_squared_error(y_test, y_pred_rf))
rf_squared = r2_score(y_test, y_pred_rf)


full_metrics = pd.Series({'Model': "Random Forest ", 'RMSE':rf_RMSE, 'R-Squared': rf_squared})

print(full_metrics)
```
